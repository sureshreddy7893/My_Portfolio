Arithmatic operation: getting values from user

```python
a=int(input("Enter the value of a:"))
b=int(input("Enter the value of b:"))
add = a+b
sub = a-b
mul = a*b
div = a//b
mod = a%b
print(add,sub,mul,div,mod)
print("sum: {0},diff: {1},Mul: {2},Div: {3},Mod: {4}".format(add,sub,mul,div,mod))
print("sum: %d,diff: %d,Mul:%d,Div:%2.3f,Mod:%d"%(add,sub,mul,div,mod))
```

< : left align, > : Right align

```python
print("{0:<4} | {1:<4} | {2:<4} | {3:>4} | {4:<4}".format("Sum","Diff","Mul","Div","Mod"))
print("{0:<4} | {1:<4} | {2:<4} | {3:<4} | {4:<4}".format(add,sub,mul,div,mod))
```

###While - Break, Continue

```python
a=0
while a<10:
  a+=1
  print(a)
  if a==5:
    print('a is equal to 5')
    break;
  else:
    print('cotinue')
    continue;
```
