```python
ds=pd.read_excel('Land_Price_Prediction.xlsx')
ds
```

```python
plt.scatter(ds.Land,ds.Price,color='red',marker='*')
plt.xlabel('Land')
plt.ylabel('Price')
plt.title("Land Price Prediction")
plt.show()
```

```python
land=ds.drop('Price',axis='columns')
land
Price=ds.Price
Price
```

```python
model=LinearRegression()
model.fit(land,Price)
```

```python
Land=[[1500]]
Prediction=model.predict(Land)
Prediction
```
