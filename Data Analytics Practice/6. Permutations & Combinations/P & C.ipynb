```python
from itertools import permutations
count=0
data=[1,2,3,4]
perm=permutations(data,2)
for i in list(perm):
  print(i)
  count+=1
print(count)
```

```python
from itertools import combinations
count=0
data=[1,2,3,4]
COMB=combinations(data,2)
for i in list(COMB):
  print(i)
  count+=1
print(count)
```

```python
from itertools import combinations_with_replacement as cr
count=0
data=[1,2,3,4]
perm=cr(data,2)
for i in list(perm):
  print(i)
  count+=1
print(count)
```

```python
import itertools
count=0
co=['H','T']
result=list(itertools.product(co,repeat=3))
for r in result:
  print(r)
  count+=1
print("Perm:",count)
```
