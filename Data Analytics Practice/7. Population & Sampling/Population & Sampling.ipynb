```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv('Dataset.csv')
print(dataset.shape)
print(dataset.describe())
print(dataset.head(5))
```

###SimpleRandomSampling

```python
simpleRandomSample = dataset.sample(n=10).sort_values(by='Id')
mean_simpleRandomSample = round(simpleRandomSample['SalePrice'].mean(),3)
print('mean_simpleRandomSample:',mean_simpleRandomSample)
print(simpleRandomSample)
```

###SystematicSampling

```python
index = np.arange(0,len(dataset),step=50)
systematicSample = dataset.iloc[index]
mean_systematicSample = round(systematicSample["SalePrice"].mean(),3)
print('mean_systematicSample:',mean_systematicSample)
print(systematicSample)
print(systematicSample.shape)
```

###ClusterSampling

```python
n=10
dataset['Cluster_Id']=np.repeat([range(1,n+1)],len(dataset)/n)
index=[]
for i in range(1,len(dataset)):
  if dataset['Cluster_Id'].iloc[i]%2==0:
    index.append(i)
clusterSample = dataset.iloc[index]
mean_clusterSample = round(clusterSample['SalePrice'].mean(),3)
print('mean_clusterSample:',mean_clusterSample)
print(clusterSample)
print(clusterSample.head(10))
```

###StratifiedSampling

```python
dataset=pd.read_csv("Dataset.csv")
dataset['strata']=np.repeat([1,2],len(dataset)/2)
index=[]
for i in range(len(dataset)):
  index.append(i)
stratifiedSample =dataset.iloc[index]
mean_stratifiedSample =round(dataset['SalePrice'].mean(),3)
print(mean_stratifiedSample)
print(stratifiedSample)
```

```python
from sklearn.model_selection import StratifiedShuffleSplit as sss
split=sss(n_splits=1,test_size=8)
for x,y in split.split(dataset,dataset['strata']):
  stratifiedRandomSample=dataset.iloc[y].sort_values(by='Id')
print(stratifiedRandomSample)
mean_stratifiedRandomSample = round(stratifiedRandomSample['SalePrice'].mean(),3)
print('mean_stratifiedRandomSample:',mean_stratifiedRandomSample)
```
