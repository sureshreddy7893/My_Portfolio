```python
from pandas import read_csv
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
URL='Energy Meter.csv'
name=['Voltage','Current','Power','Class']
ds=read_csv(URL,names=name)
ds
```

```python
array= ds.values
X=array[:,0:3]
Y=array[:,3]
X_train,X_validation,Y_train,Y_validation=train_test_split(X,Y,test_size=0.50,random_state=1)
model=SVC(gamma='auto')
```

```python
model.fit(X_train,Y_train)
```

```python
import pickle
filename='model.pkl'
pickle.dump(model,open(filename,'wb'))
```

```python
loded_model=pickle.load(open(filename,'rb'))
result=loded_model.score(X_validation,Y_validation)
print(result)
```

```python
value=[[214.6,0.24,52.9]]
predictions=model.predict(value)
print(predictions)
```
