```python
def factorial(n):
  result=1
  for i in range(1,n+1):
    result *=i
  return result
print(factorial(4))
```

```python
def arith(a,b):
  sum=a+b
  dif=a-b
  mul=a*b
  return sum,dif,mul
var1=int(input("Enter 1st number:"))
var2=int(input("Enter 2nd number:"))
result=arith(var1,var2)
print(result)
print("difference:",result[1])
```

```python
def factorial(n):
	result=1
	for i in range(1,n+1):
		result*=i
	return result
print(factorial(5))
```

```python
def factorial(n):
	if n==0 or n==1:
		return 1
	else:
		return n*factorial(n-1)
print(factorial(5))
```
